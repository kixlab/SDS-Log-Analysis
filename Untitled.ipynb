```python
dirs = ['plotly_export', 'chartblocks', 'd3_clean', 'fusion_clean', 'graphiq_clean']

with open('labels.json') as f:
    labels = json.load(f)
    
print(labels)
```

```python
data = []
def fix_id(obj, type):
    obj['type'] = type
    
    return obj
    
for dir in dirs:
    with open('./%s/urls.txt' % dir) as g:
        charts = csv.DictReader(g, delimiter = ' ', fieldnames = ['id', 'url'], restkey='chart_type')
        
        d = [fix_id(c, dir) for c in charts]
        
        data.extend(d)

print(d)

```

```python
def parse_label(chart_obj):
    txt = chart_obj['chart_type'][0]
    if txt in labels.keys():
        chart_obj['parsed_type'] = labels[txt]
    elif ',' in txt:
        splitted = txt.split(',')[0]
        chart_obj['parsed_type'] = labels[splitted] if splitted in labels.keys() else splitted
    else:
        chart_obj['parsed_type'] = txt
    return chart_obj
        

filtered_data = [parse_label(d) for d in data]
filtered_data = [d for d in filtered_data if not d['parsed_type'].isnumeric()]
```
